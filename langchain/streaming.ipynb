```python
# pyright: reportArgumentType=false

import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain.agents import create_agent
from langchain.messages import HumanMessage
from rich import print as rprint


load_dotenv()


model = ChatOpenAI(
    model=os.environ.get("MODEL_NAME"),
    temperature=0,
    base_url=os.environ.get("COMPATIBLE_BASE_URL"),
    api_key=os.environ.get("COMPATIBLE_API_KEY"),
)


def get_weather(city: str) -> str:
    """Get weather for a given city."""

    return f"It's always sunny in {city}!"


agent = create_agent(
    model=model,
    tools=[get_weather],
)

input = {"messages": [HumanMessage("What is the weather in SF?")]}

agent.invoke(input)
```

# stream_mode="values"

```python
# pyright: reportArgumentType=false

for chunk in agent.stream(input, stream_mode="values"):
    rprint(chunk)
    print("============")
```

```python
# pyright: reportArgumentType=false

for chunk in agent.stream(input, stream_mode="values"):
    chunk["messages"][-1].pretty_print()
```

# stream_mode="updates"

```python
# pyright: reportArgumentType=false

for chunk in agent.stream(input, stream_mode="updates"):
    for step, content in chunk.items():
        rprint(f"step: {step} \ncontent: {content}")
    print("============")
```

# stream_mode="messages"

```python
# pyright: reportArgumentType=false

for chunk in agent.stream(input, stream_mode="messages"):
    rprint(chunk)
    print("============")
```

```python
# pyright: reportArgumentType=false
from langchain.messages import AIMessageChunk


for msg, metadata in agent.stream(input, stream_mode="messages"):
    if isinstance(msg, AIMessageChunk) and msg.content:
        print(msg.content)
```

# stream_mode="custom"

```python
# pyright: reportArgumentType=false

from langgraph.config import get_stream_writer


def get_weather(city: str) -> str:
    """Get weather for a given city."""

    writer = get_stream_writer()
    # stream any arbitrary data
    writer(f"Looking up data for city: {city}")
    writer(f"Acquired data for city: {city}")
    return f"It's always sunny in {city}!"


agent = create_agent(
    model=model,
    tools=[get_weather],
)

for chunk in agent.stream(input, stream_mode="custom"):
    rprint(chunk)
```

# ✅ 5. stream_mode=[...]

```python
# pyright: reportArgumentType=false

for stream_mode, event in agent.stream(
    input, stream_mode=["values", "updates", "messages", "custom"]
):
    rprint(f"stream_mode: {stream_mode}\nevent: {event}\ntype of event: {type(event)}")
    print("============")
```
